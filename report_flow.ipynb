```python
def get_day_spanning(df, tv_cols, contain_pre_day=True, title='Title', desc='Description',
                     start='Program Start time', dur='Duration', end='Program End Time',
                     date='Date', weekday='Weekday', channel='Channel'):
    
    # 读取映射表
    mapping = pd.read_csv("D:/wangxiaoyang/Regular_Work/support_files/channel_mapping.csv")
    
    # 第一步：创建新列，基于传入的列名
    df['Title'] = df[title]
    df['Description'] = df[desc]
    df['Channel'] = df[channel]
    df['Date'] = pd.to_datetime(df[date])
    df['Weekday'] = df[weekday]
    df['Start'] = df[start]
    df['Dur'] = df[dur]
    df['End'] = df[end]
    
    # 第二步：过滤掉特定的行，处理 "26:00:00" 和 "2:00:00"
    df_normal = df[(df['End'] != "26:00:00") & (df['Start'] != "2:00:00")].copy()
    
    # 使用 .iloc 选择 tv_cols 代表的列
    cols = [df.columns.get_loc(c) for c in ['Regions', 'Title', 'Description', 'Channel', 'Date', 'Weekday', 'Start', 'End', 'Dur']]
    cols.extend(tv_cols)
    df_normal = df_normal.iloc[:, cols]
    
    # 如果 `contain_pre_day` 为 True，则过滤掉最小日期的数据
    if contain_pre_day:
        df_normal = df_normal[df_normal['Date'] != df_normal['Date'].min()]
    
    # 第三步：处理特殊情况，筛选时间为 "26:00:00" 或 "2:00:00" 的行
    df_special = df[(df['End'] == "26:00:00") | (df['Start'] == "2:00:00")].copy()
    
    # 如果只有1行或0行，直接返回结果
    if len(df_special) <= 1:
        return [df_normal]
    
    # 第四步：对特殊情况进行排序和进一步处理
    df_special = df_special.sort_values(by=['Channel', 'Date', 'Start'])
    
    df_special['indicator'] = np.where(df_special['Start'] == "2:00:00", 2, 
                                       np.where(df_special['End'] == "26:00:00", 1, 0))
    
    df_special['Date_lead'] = df_special['Date'].shift(-1)
    df_special['Date_next'] = df_special['Date'] + timedelta(days=1)
    df_special['Title_lead'] = df_special['Title'].shift(-1)
    
    df_special['day_span'] = (df_special['Title'] == df_special['Title_lead']) & (df_special['Date_lead'] == df_special['Date_next'])
    df_special.reset_index(drop=True)
    
    # 第五步：检查哪些行需要处理
    rows_to_check = []
    name_list = []
    
    for i in range(len(df_special)):
        if df_special.iloc[i]['day_span']:
            rows_to_check.extend([i, i+1])
            name_list.extend([f"{df_special.iloc[i]['Date']} {df_special.iloc[i]['Channel']}"] * 2)
    
    # 第六步：与映射数据合并并格式化结果
    res_df = df_special.iloc[rows_to_check].merge(mapping, how='left', left_on='Channel', right_on=channel)
    
    # 调整 End 时间并格式化结果文本
    res_df['End'] = np.where(res_df['End'] == "26:00:00", "25:59:59", res_df['End'])
    
    res = res_df.apply(
        lambda row: f"{row['Date'].strftime('%Y%m%d')} {row['code']:04} "
                    f"{row['Start'].replace(':', '')} {row['End'].replace(':', '')} {row['Channel']}",
        axis=1
    ).tolist()
    
    # 第七步：调整持续时间并合并 DataFrames
    df_ds = df_special.iloc[rows_to_check].copy()
    
    for i in range(0, len(df_ds), 2):
        # 调整 End 时间
        df_ds.at[i, 'End'] = df_ds.at[i+1, 'End']
        
        # 调整持续时间
        dur_1 = pd.to_timedelta(df_ds.iloc[i]['Dur'])
        dur_2 = pd.to_timedelta(df_ds.iloc[i+1]['Dur'])
        total_duration = dur_1 + dur_2
        df_ds.at[i, 'Dur'] = str(total_duration)
    
    # 删除跨天的数据
    df_normal = df_special.drop(index=rows_to_check)
    
    # 返回最终结果
    return ['\n'.join(res),
        pd.concat([df_normal, df_ds]).drop_duplicates(),
        df_ds.iloc[::2]  # 每隔一行取一次
           ]
```

```python
# 转换结束时间和时长
def convert_time(seconds, excel=True):
    # 将秒数转换为 hh:mm:ss 格式
    hours = seconds // 3600
    minutes = (seconds % 3600) // 60
    seconds = seconds % 60
    return f"{hours}:{minutes:02}:{seconds:02}"

def get_seconds(time_str, durr = True):
    # 将 hh:mm:ss 转换为秒数
    h, m, s = map(int, time_str.split(':'))
    ss = h * 3600 + m * 60 + s
    
    if not durr:
        if ss < 2*3600 + 5*60:
            ss += 24*3600
    return ss

cdp_epl['结束时间'] = np.where(
    cdp_epl['时长'] == "1:25:00",
    cdp_epl['结束时间'].apply(lambda x: convert_time(get_seconds(x, durr=False) - 5 * 60, excel=False)),
    cdp_epl['结束时间']
)

cdp_epl['时长'] = np.where(
    cdp_epl['时长'] == "1:25:00", 
    "1:20:00", 
    cdp_epl['时长']
)

# 选择需要的列
cdp_epl = cdp_epl[['Regions', 'Program', '日期', 'Weekday', '开始时间', '结束时间', '时长', '频道', 'Program', 'Description']]
```
